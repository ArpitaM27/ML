```python
#This is project 1 of basic python where we ask the user to give their expenditure output and our calculator will help them add their expenditure.
total=0;
exp=1
maxexp=0
minexp=None
while exp!=0:
  exp=int(input("Enter your expenditure(enter 0 to stop): "));
  if exp==0:
      break
  total=total+exp;

  if exp>maxexp:
    maxexp=exp
  if minexp==None or exp<minexp:
    minexp=exp
print("your total expenditure is: ",total)
print("your maximum expenditure is: ",maxexp)
print("your minimum expenditure is: ",minexp)
```
