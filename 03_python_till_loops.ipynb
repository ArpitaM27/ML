```python
#search for no x in tuple
no=(1,3,5,7,9,11,23)
x=int(input("enter no to be searched: "))
index=0
while(index<=(len(no)-1)):
  if(x==no[index]):
    print("found at position ",index+1)
    break
  index+=1

else:
  print("not found")
```
